```python
%%time
import pandas as pd
```

```python
cleaned_df = pd.read_csv('cleaned_playerbox.csv')
```

```python
list_players = list(cleaned_df['playDispNm'].unique())
```

```python
cleaned_df[:3]
```

```python
%%time
pre_series_list = pd.DataFrame()
pre_merged_list = pd.DataFrame()
for player in list_players:
    date_array = cleaned_df[cleaned_df['playDispNm'] == player].gmDate.to_numpy()
    second_series = pd.Series(date_array, name = 'gmDate')
    idx = list(range(1, len(date_array)+1))
    first_series = pd.Series(date_array, index = idx, name = 'pre_gmDate')
    pre_series_list = pd.merge(first_series, second_series, left_index=True, right_index=True)
    pre_series_list['playDispNm'] = player
    pre_merged_list = pd.concat([pre_series_list,pre_merged_list])
```

```python
pre_merged_list[:10]
```

```python
%%time
series_list = pd.DataFrame()
merged_list = pd.DataFrame()
for player in list_players:
    date_array = cleaned_df[cleaned_df['playDispNm'] == player].gmDate.to_numpy()
    second_series = pd.Series(date_array, name = 'next_gmDate')
    idx = list(range(1, len(date_array)+1))
    first_series = pd.Series(date_array, index = idx, name = 'gmDate')
    series_list = pd.merge(first_series, second_series, left_index=True, right_index=True)
    series_list['playDispNm'] = player
    merged_list = pd.concat([series_list[1:],merged_list])
```

```python
merged_list[:10]
```

```python
cleaned_df_new = pd.merge(pre_merged_list, merged_list, on=['playDispNm', 'gmDate'])
```

```python
cleaned_df_new[:10]
```

```python
cleaned_df_new['gmDate'] = pd.to_datetime(cleaned_df_new['gmDate'])
```

```python
cleaned_df_new['next_gmDate'] = pd.to_datetime(cleaned_df_new['next_gmDate'])
```

```python
cleaned_df_new['pre_gmDate'] = pd.to_datetime(cleaned_df_new['pre_gmDate'])
```

```python
cleaned_df_new['next_gmDays'] = cleaned_df_new['next_gmDate'] - cleaned_df_new['gmDate']
cleaned_df_new['pre_gmDays'] = cleaned_df_new['gmDate'] - cleaned_df_new['pre_gmDate']
```
